```python
all_df_copy = all_df.copy()
all_df_copy.set_index('MONTHLY REPORTING PERIOD', inplace=True)

# Check for missing values
all_df_copy = all_df_copy.dropna()
for lag in range(1, 13):  # Lags for past 12 months
    all_df_copy[f'delinquency_lag_{lag}'] = all_df_copy['CURRENT LOAN DELINQUENCY STATUS'].shift(lag)

all_df_copy['rolling_mean_3'] = all_df_copy['CURRENT LOAN DELINQUENCY STATUS'].rolling(3).mean()
all_df_copy['rolling_std_3'] = all_df_copy['CURRENT LOAN DELINQUENCY STATUS'].rolling(3).std()

all_df_copy.dropna(inplace=True)
```

```python
train = all_df_copy.iloc[:-48]  # last 12 time points as test
test = all_df_copy.iloc[-48:]

X_train = train.drop(columns=['CURRENT LOAN DELINQUENCY STATUS'])
y_train = train['CURRENT LOAN DELINQUENCY STATUS']
X_test = test.drop(columns=['CURRENT LOAN DELINQUENCY STATUS'])
y_test = test['CURRENT LOAN DELINQUENCY STATUS']
```

```python
from lightgbm import LGBMRegressor

model = LGBMRegressor()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

print('MAE:', mean_absolute_error(y_test, y_pred))
print('RMSE:', mean_squared_error(y_test, y_pred, squared=False))
```

```python
plt.plot(y_test.index, y_test, label='Actual')
plt.plot(y_test.index, y_pred, label='Forecast')
plt.legend()
plt.title("Delinquency Forecast vs Actual")
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense


target_col = 'CURRENT LOAN DELINQUENCY STATUS'
features = all_df_copy.columns.tolist()

# Normalize all features
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(all_df_copy), columns=features, index=all_df_copy.index)

# --- Convert to supervised format ---
def create_sequences(data, target, window=12):
    X, y = [], []
    for i in range(window, len(data)):
        X.append(data[i - window:i])
        y.append(target[i])
    return np.array(X), np.array(y)

window_size = 12
X, y = create_sequences(df_scaled.values, df_scaled[target_col].values, window=window_size)

# Split
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# --- Build LSTM Model ---
model = Sequential([
    LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=False),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=30, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# --- Predict ---
y_pred = model.predict(X_test)

# Inverse scale if needed
test_dates = all_df_copy.index[window_size + split : window_size + split + len(y_test)]

plt.figure(figsize=(12, 6))
plt.plot(test_dates, y_test, label='Actual')
plt.plot(test_dates, y_pred.flatten(), label='Predicted')
plt.xlabel('Date')
plt.ylabel('CURRENT LOAN DELINQUENCY STATUS')
plt.title('LSTM Forecast vs Actual')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import torch
from pytorch_forecasting import TimeSeriesDataSet, TemporalFusionTransformer, Baseline
from pytorch_forecasting.data import GroupNormalizer
from pytorch_forecasting.metrics import RMSE
from pytorch_lightning import seed_everything, Trainer

# --- Data ---
all_df_copy['time_idx'] = range(len(all_df_copy))  # convert date to time_idx
all_df_copy['group'] = "series_1"  # single series

# --- Dataset ---
max_encoder_length = 12
max_prediction_length = 3

dataset = TimeSeriesDataSet(
    all_df_copy,
    time_idx='time_idx',
    target='CURRENT LOAN DELINQUENCY STATUS',
    group_ids=['group'],
    max_encoder_length=max_encoder_length,
    max_prediction_length=max_prediction_length,
    time_varying_unknown_reals=['CURRENT LOAN DELINQUENCY STATUS'],
    time_varying_known_reals=['time_idx'],
    time_varying_known_categoricals=['group'],
    target_normalizer=GroupNormalizer(groups=['group']),
)

# Split
cutoff = int(all_df_copy["time_idx"].max() * 0.8)

# Split the DataFrame
training_df = all_df_copy[all_df_copy["time_idx"] <= cutoff]
validation_df = all_df_copy[all_df_copy["time_idx"] > cutoff]

# Create datasets
train = TimeSeriesDataSet.from_dataset(dataset, training_df)
val = TimeSeriesDataSet.from_dataset(dataset, validation_df)

# --- Dataloaders ---
from torch.utils.data import DataLoader
batch_size = 64
train_loader = DataLoader(train, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val, batch_size=batch_size)

# --- Model ---
tft = TemporalFusionTransformer.from_dataset(
    dataset,
    learning_rate=0.03,
    hidden_size=16,
    attention_head_size=1,
    dropout=0.1,
    hidden_continuous_size=8,
    output_size=1,
    loss=RMSE(),
    log_interval=10,
    reduce_on_plateau_patience=4,
)

print(type(tft))
# --- Train ---

trainer = Trainer(
    max_epochs=30,
    gradient_clip_val=0.1,
    enable_checkpointing=False  # disable warnings
)

trainer.fit(
    model=tft,
    train_dataloaders=train_loader,
    val_dataloaders=val_loader
)

# --- Forecast ---
raw_predictions, x = tft.predict(val, mode='raw', return_x=True)

# --- Plot ---
tft.plot_prediction(x, raw_predictions, idx=0)
```
